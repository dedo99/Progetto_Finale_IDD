# CREAZIONE SCHEMA MEDIATO

```python
import os
import pandas as pd
from valentine import valentine_match, valentine_metrics
from valentine.algorithms import Coma, DistributionBased
import pprint
```

```python
# ---------------------------------------------------------
# CREATA PER INIZIALIZZARE UN DATASET FITTIZIO MA ORA NON
# SI UTILIZZA IL DATASET FITTIZIO DA NOI CREATO MANUALMENTE
# ---------------------------------------------------------
def creation_dataset_confronto(df1):
  # Load data using pandas
  df2 = pd.read_json("datasets/output_disfold.json")

  matcher_coma = Coma(strategy="COMA_OPT")
  matches_coma = valentine_match(df1, df2, matcher_coma)

  attributi_mediato = []
  attributi_dataset = []
  for i in matches_coma:
    if matches_coma[i] > 0.5:
      attributi_mediato.append(i[0][1])
      attributi_dataset.append(i[1][1])

  for i in df2.index:
      data = {}
      # print(i)
      for d,m in zip(attributi_dataset,attributi_mediato):
        # print(df2[d][i])
        data[m] = df2[d][i]
      temp_data=pd.DataFrame([data])
      df1=pd.concat([df1, temp_data], ignore_index=True)

  return df1
```

```python
def creation_df_log_matcher(df_logmachers_temp, matches, i, type, name_dataset):
  record = {}
  record['tab1'] = i[0][0]
  record['attr1'] = i[0][1]
  record['tab2'] = name_dataset
  record['attr2'] = i[1][1]
  record['prob'] = matches[i]
  if type == 'coma':
    record['matcher'] = 'Coma'
  else:
    record['matcher'] = 'DistributionBased'
  temp_record=pd.DataFrame([record])
  # df_logmachers=df_logmachers.append(record, ignore_index=True)
  df_logmachers_temp=pd.concat([df_logmachers_temp, temp_record], ignore_index=True)
  return df_logmachers_temp
```

```python
def creation_schema_alignment(df1, df2, df_confronto, df_logmachers, name_dataset):

  # Instantiate matcher and run
  # Coma requires java to be installed on your machine
  # If java is not an option, all the other algorithms are in Python (e.g., Cupid)
  matcher_coma = Coma(strategy="COMA_OPT")
  matches_coma = valentine_match(df1, df2, matcher_coma)

  distr_run = False

  df_logmachers_temp = pd.DataFrame(columns = ['tab1' , 'attr1', 'tab2', 'attr2', 'prob', 'matcher'])
 
  attributi_mediato = []
  attributi_dataset = []
  for i in matches_coma:
    df_logmachers_temp = creation_df_log_matcher(df_logmachers_temp, matches_coma, i, 'coma', name_dataset)
    if matches_coma[i] > 0.5:
      attributi_mediato.append(i[0][1])
      attributi_dataset.append(i[1][1])
    else:
      try:
        if not distr_run:
          matcher_distr = DistributionBased()
          matches_distr = valentine_match(df_confronto, df2, matcher_distr)
          distr_run = True
        # distribution based già instanziato
        # log del matcher su Distribution Based
        df_logmachers_temp = creation_df_log_matcher(df_logmachers_temp, matches_distr, i, 'distribution', name_dataset)
        # se nel Distribution Based c'è la stessa coppia di attributi 
        # che fa match con valore superiore a 0.5 allora fa match
        if matches_distr[i] > 0.5:
          attributi_mediato.append(i[0][1])
          attributi_dataset.append(i[1][1])
      except:
        pass
  # print(attributi_dataset)
  # print(attributi_mediato)

  df1_temp = pd.DataFrame(columns = ['name' , 'industry', 'market_cap', 'employees', 'country', 'headquarters', 'address', 'sector', 'ceo', 'founded', 'founders', 'share_price', 'website', 'stock', 'area_served', 'revenue'])

  # inserimento dei dati nel dataframe
  for i in df2.index:
    data = {}
    # print(i)
    for d,m in zip(attributi_dataset,attributi_mediato):
      # print(df2[d][i])
      data[m] = df2[d][i]
    temp_data=pd.DataFrame([data])
    df1_temp=pd.concat([df1_temp, temp_data], ignore_index=True)

  pp = pprint.PrettyPrinter(indent=4)

  # stampa dei matchers
  pp.pprint(matches_coma)
  if distr_run:
    pp.pprint(matches_distr)


  return df1_temp, df_logmachers_temp
  # print(df1)
```

```python
#funzione utilizzata per trasformare gli elementi del dataframe che sono una lista in stringa 
def list2string(x):
    result = ''
    if(isinstance(x, list)):
        for element in x:
            result = result + ', ' + str(element)
        return result[1:]
    else:
        return x
```

```python
from threading import Thread
import time
class IlMioThread (Thread):

    def __init__(self, df1, df2, df_confronto, df_logmachers, name_dataset):
        Thread.__init__(self)
        self.df1 = df1
        self.df2 = df2
        self.df_confronto = df_confronto
        self.df_logmachers = df_logmachers
        self.name_dataset = name_dataset
   
    def run(self):
        try:
            self.df1, self.df_logmachers = creation_schema_alignment(self.df1, self.df2, self.df_confronto, self.df_logmachers, self.name_dataset)
        except:
            print("ERRORE " + str(self.name_dataset))
        
```

```python
from pathlib import Path
from itertools import zip_longest
import time
#formati json, csv, xls, xlsx, jsonl
df1 = pd.DataFrame(columns = ['name' , 'industry', 'market_cap', 'employees', 'country', 'headquarters', 'address', 'sector', 'ceo', 'founded', 'founders', 'share_price', 'website', 'stock', 'area_served', 'revenue'])
# df_confronto = creation_dataset_confronto(df1)
df_confronto = pd.read_json('output_alignment_multithread/schema_mediato.json')
df_logmachers = pd.DataFrame(columns = ['tab1' , 'attr1', 'tab2', 'attr2', 'prob', 'matcher'])


df1_a = df1
df1_b = df1
df1_c = df1
df1_d = df1
df1_e = df1

df_logmachers_a = df_logmachers
df_logmachers_b = df_logmachers
df_logmachers_c = df_logmachers
df_logmachers_d = df_logmachers
df_logmachers_e = df_logmachers


#numero di errori nell'esecuzione di file
num_error = 0

directory = 'datasets'
dataset_list_json = Path(directory).glob('*.json')
dataset_list_jsonl = Path(directory).glob('*.jsonl')
dataset_list_csv = Path(directory).glob('*.csv')
dataset_list_xls = Path(directory).glob('*.xls')
dataset_list_xlsx = Path(directory).glob('*.xlsx')
i = 0

start_time = time.time()


for a,b,c,d,e in zip_longest(dataset_list_json, dataset_list_jsonl, dataset_list_csv, dataset_list_xls, dataset_list_xlsx, fillvalue='?'):

  df1_a = df1
  df1_b = df1
  df1_c = df1
  df1_d = df1
  df1_e = df1

  df_logmachers_a = df_logmachers
  df_logmachers_b = df_logmachers
  df_logmachers_c = df_logmachers
  df_logmachers_d = df_logmachers
  df_logmachers_e = df_logmachers


  if a != '?':
    print(str(i) + " " + str(a))
    i=i+1
    df2_a = pd.read_json(a)
    # trasforma le celle del dataframe che hanno liste in stringhe
    df2_a = df2_a.applymap(list2string)
    name_dataset_a = str(a).split('\\')[1]
    thread1 = IlMioThread(df1_a, df2_a, df_confronto, df_logmachers_a, name_dataset_a)
    thread1.start()
  else:
    thread1 = None
  
  if b != '?':
    print(str(i) + " " + str(b))
    i=i+1
    df2_b = pd.read_json(b, lines=True)
    # trasforma le celle del dataframe che hanno liste in stringhe
    df2_b = df2_b.applymap(list2string)
    name_dataset_b = str(b).split('\\')[1]
    thread2 = IlMioThread(df1_b, df2_b, df_confronto, df_logmachers_b, name_dataset_b)
    thread2.start()
  else:
    thread2 = None

  
  if c != '?':
    print(str(i) + " " + str(c))
    i=i+1
    df2_c = pd.read_csv(c, encoding='ISO-8859-1')
    # trasforma le celle del dataframe che hanno liste in stringhe
    df2_c = df2_c.applymap(list2string)
    name_dataset_c = str(c).split('\\')[1]
    thread3 = IlMioThread(df1_c, df2_c, df_confronto, df_logmachers_c, name_dataset_c)
    thread3.start()
  else:
    thread3 = None

  if d != '?':
    print(str(i) + " " + str(d))
    i=i+1
    df2_d = pd.read_excel(d)
    # trasforma le celle del dataframe che hanno liste in stringhe
    df2_d = df2_d.applymap(list2string)
    name_dataset_d = str(d).split('\\')[1]
    thread4 = IlMioThread(df1_d, df2_d, df_confronto, df_logmachers_d, name_dataset_d)
    thread4.start()
  else:
    thread4 = None

  if e != '?':
    print(str(i) + " " + str(e))
    i=i+1
    df2_e = pd.read_excel(e)
    # trasforma le celle del dataframe che hanno liste in stringhe
    df2_e = df2_e.applymap(list2string)
    name_dataset_e = str(e).split('\\')[1]
    thread5 = IlMioThread(df1_e, df2_e, df_confronto, df_logmachers_e, name_dataset_e)
    thread5.start()
  else:
    thread5 = None


  if thread1 != None:
    thread1.join()
  if thread2 != None:
    thread2.join()
  if thread3 != None:
    thread3.join()
  if thread4 != None:
    thread4.join()
  if thread5 != None:
    thread5.join()

  if thread1 != None:
    df1=pd.concat([df1, thread1.df1], ignore_index=True)
  if thread2 != None:
    df1=pd.concat([df1, thread2.df1], ignore_index=True)
  if thread3 != None:
    df1=pd.concat([df1, thread3.df1], ignore_index=True)
  if thread4 != None:
    df1=pd.concat([df1, thread4.df1], ignore_index=True)
  if thread5 != None:
    df1=pd.concat([df1, thread5.df1], ignore_index=True)
  
  
  if thread1 != None:
    df_logmachers=pd.concat([df_logmachers, thread1.df_logmachers], ignore_index=True)
  if thread2 != None:
    df_logmachers=pd.concat([df_logmachers, thread2.df_logmachers], ignore_index=True)
  if thread3 != None:
    df_logmachers=pd.concat([df_logmachers, thread3.df_logmachers], ignore_index=True)
  if thread4 != None:
    df_logmachers=pd.concat([df_logmachers, thread4.df_logmachers], ignore_index=True)
  if thread5 != None:
    df_logmachers=pd.concat([df_logmachers, thread5.df_logmachers], ignore_index=True)
  

  
end_time = time.time()
total_time = end_time - start_time

display(df1)
display(df_confronto)
display(df_logmachers)
print("Tempo Esecuzione: " + str(total_time))
```

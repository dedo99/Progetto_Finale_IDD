# Record linkage con Python Record Linkage Toolkit

```python
import recordlinkage
import pandas as pd
```

```python
df = pd.read_json('output_alignment_multithread_finalpass/schema_mediato.json')
print(len(df))

# half_size = int(len(df)/2)
# df_a = df.iloc[:half_size,:]
# df_b = df.iloc[half_size:,:]

df_a = df
df_b = df

display(df_a)
display(df_b)
```

```python
print(df[df.index == 1202]['name'])
print(df[df.index == 1202]['name'].values[0])
```

```python
indexer = recordlinkage.Index()
indexer.block('name')
candidate_links = indexer.index(df_a, df_b)

print(candidate_links)
```

```python
set_no_copy = set()
for (c,b) in candidate_links:
    if (b,c) not in set_no_copy and (c != b):
        set_no_copy.add((c,b))

print(len(set_no_copy))

list_no_copy = list(set_no_copy)
# gli elementi delle coppie vengono distribuite su due liste parallele
list_0 = [x[0] for x in list_no_copy]
list_1 = [x[1] for x in list_no_copy]
```

```python
multi_index = pd.MultiIndex.from_arrays([list_0, list_1])
```

```python
print(multi_index)
```

# Unione di df_match e df_no_match e suddivisione in train, test, validation per Deepmatcher

```python
df_match = pd.read_csv('output_match/df_match.csv')
df_no_match = pd.read_csv('output_match/df_no_match.csv')
df_match = df_match.sample(len(df_no_match))

df_match = df_match.drop(['Unnamed: 0'], axis = 1)
df_no_match = df_no_match.drop(['Unnamed: 0'], axis = 1)
```

```python
df_union = pd.concat([df_match, df_no_match], ignore_index=True)
# display(df_union)

```

```python
df_train = df_union.sample(7000)
df_test = df_union.sample(2300)
df_validation = df_union.sample(2300)

# Crea un nuovo attributo "id" come indice numerico progressivo per TRAIN SET
df_train['id'] = range(1, len(df_train) + 1)
df_train = df_train.reset_index(drop=True)
# Imposta la colonna "id" come prima colonna
cols = df_train.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_train = df_train[cols]

# Crea un nuovo attributo "id" come indice numerico progressivo per TEST SET
df_test['id'] = range(1, len(df_test) + 1)
df_test = df_test.reset_index(drop=True)
# Imposta la colonna "id" come prima colonna
cols = df_test.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_test = df_test[cols]

# Crea un nuovo attributo "id" come indice numerico progressivo per VALIDATION SET
df_validation['id'] = range(1, len(df_validation) + 1)
df_validation = df_validation.reset_index(drop=True)
# Imposta la colonna "id" come prima colonna
cols = df_validation.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_validation = df_validation[cols]


display(df_train)
display(df_test)
display(df_validation)
```

```python
df_train.to_csv('output_match/df_train.csv')
df_test.to_csv('output_match/df_test.csv')
df_validation.to_csv('output_match/df_validation.csv')
```

```python
import deepmatcher as dm
train, validation, test = dm.data.process(path='output_match',
    train='df_train.csv', validation='df_validation.csv', test='df_test.csv')
```

```python
model = dm.MatchingModel()
```

```python
model.run_train(train, validation, best_save_path='best_model.pth')
```

```python
model.run_eval(test)

unlabeled = dm.data.process_unlabeled(path='data_directory/unlabeled.csv', trained_model=model)
model.run_prediction(unlabeled)
```
